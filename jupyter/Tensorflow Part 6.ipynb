```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('TF_2_Notebooks_and_Data/DATA/RSCCASN.csv',parse_dates=True,index_col='DATE')
df.colums = ['Sales']
df
```

```python
df.plot(figsize=(15,8))
```

```python
# Use 18 months for the period
test_size = 18
test_index = len(df) - test_size
train = df.iloc[:test_index]
test = df.iloc[test_index:]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(train)
scaled_train, scaled_test = scaler.transform(train), scaler.transform(test)
```

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
```

```python
length = 12
generator = TimeseriesGenerator(scaled_train, scaled_train,
                               length=length,batch_size=1)
```

```python
generator[0]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
```

```python
n_features = 1
model = Sequential()
model.add(LSTM(100, activation='relu', input_shape=(length,n_features)))
model.add(Dense(1))
model.compile(optimizer='adam',loss='mse')
```

```python
model.summary()
```

```python
(len(df)*100*18/12) + (100*18/12)
```

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stop = EarlyStopping(monitor='val_loss', patience=2)
```

```python
validation_generator = TimeseriesGenerator(scaled_test, scaled_test,
                               length=length,batch_size=1)
```

```python
model.fit_generator(generator, epochs=20, 
                    validation_data=validation_generator,
                    callbacks=[early_stop])
```

```python
pd.DataFrame(model.history.history).plot()
```

```python
test_predictions = []
first_eval_batch = scaled_train[-length:]
current_batch = first_eval_batch.reshape((1,length,n_features))

for i in range(len(scaled_test)):
    curr_pred = model.predict(current_batch)[0]
    test_predictions.append(curr_pred)
    current_batch = np.append(current_batch[:,1:,:],[[curr_pred]],axis=1)
    
true_predictions = scaler.inverse_transform(test_predictions)
test['Predictions'] = true_predictions
test
```

```python
test.plot(figsize=(12,8))
```

## Forecasting

```python
# Note: forecasting on the full data set prevents the ability to use early stopping without data leakage/cheating
# Hence, use the visualization to determine a feasible number of epochs
```

```python
full_scaler = MinMaxScaler()
scaled_full = full_scaler.fit_transform(df)
full_generator = TimeseriesGenerator(scaled_full, scaled_full,
                               length=length,batch_size=1)
```

```python
model = Sequential()
model.add(LSTM(100,activation='relu',input_shape=(length,1)))
model.add(Dense(1))
model.compile(optimizer='adam',loss='mse')
model.summary()
```

```python
model.fit_generator(full_generator,epochs=8)
```

```python
pd.DataFrame(model.history.history).plot(figsize=(15,8))
```

```python
from dateutil.relativedelta import relativedelta

forecasts = []
first_eval_batch = scaled_full[-length:]
current_batch = first_eval_batch.reshape((1,length,n_features))

for i in range(length):
    curr_pred = model.predict(current_batch)[0]
    forecasts.append(curr_pred)
    current_batch = np.append(current_batch[:,1:,:],[[curr_pred]],axis=1)
    
true_forecasts = scaler.inverse_transform(forecasts)
forecast_index = pd.date_range(df.index[-1], 
                              periods = length,
                              freq='MS')
forecast_df = pd.DataFrame(true_forecasts, index=forecast_index, columns=['Forecast'])
ax=test.plot(figsize=(15,8))
forecast_df.plot(ax=ax)
```
