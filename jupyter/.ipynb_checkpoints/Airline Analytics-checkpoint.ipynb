### Get Summary of Data

### Resample the data

```python
#Resample the data
data['sentiment']=data['airline_sentiment'].map({'positive':1,'neutral':0,'negative':-1}) 
actual_positive_sentiments = data[data['sentiment']==1].resample('10min').count()
actual_negative_sentiments = data[data['sentiment']==0].resample('10min').count()
actual_neutral_sentiments = data[data['sentiment']==-1].resample('10min').count()
```

### Standardize the data

```python
# Standardize the data
actual_positive_sentiments[['sentiment_normalized']] = \
    StandardScaler().fit_transform(actual_positive_sentiments[['sentiment']])
actual_negative_sentiments[['sentiment_normalized']] = \
    StandardScaler().fit_transform(actual_negative_sentiments[['sentiment']])
actual_neutral_sentiments[['sentiment_normalized']] = \
    StandardScaler().fit_transform(actual_neutral_sentiments[['sentiment']])
```

### Test for stationarity

```python
adfuller(actual_positive_sentiments['sentiment_normalized'])
```

p-value is less than 0.05% - should imply stationarity
