## Feature Engineering

```python
# df = df.drop(columns="id")
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].apply(lambda date: date.year)
df['month'] = df['date'].apply(lambda date: date.month)
df.head()
```

```python
plt.figure(figsize=(14,6))
sns.boxplot(x='month',y='price',data=df)
```

```python
fig, (ax1, ax2) = plt.subplots(1,2,figsize=(15,6))
ax1.set_title('By Month')
ax2.set_title('By Year')
ax1.plot(df.groupby('month').mean()['price'])
ax2.plot(df.groupby('year').mean()['price'])
```

## Using Keras

```python
df = df.drop('date',axis=1)
```

```python
X = df.drop('price',axis=1).values
y = df['price'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(22, activation='relu'))
model.add(Dense(22, activation='relu'))
model.add(Dense(22, activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam',loss='mse')
model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),
          batch_size=128,
         epochs=500)
```

```python
fig = plt.figure(figsize=(14,6))
plt.plot(pd.DataFrame(model.history.history))
```

```python
#!jt -r
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test,predictions))
```

```python
explained_variance_score(y_test,predictions)
```

```python
plt.figure(figsize=(14,6))
plt.scatter(y_test,predictions)
plt.plot(y_test,y_test,'r')
```

```python
single_house = df.drop('price',axis=1).iloc[0]
single_house = single_house.values.reshape(-1,single_house.shape[0])
single_house = scaler.transform(single_house)
model.predict(single_house)
```
